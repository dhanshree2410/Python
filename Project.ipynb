```python

# Restaurant Bill using Dictionary

menu = {"Tea":20,"coffee":40,"Cold Coffee": 80,"Pizza": 200,"Burger": 120,"Pasta": 150}
order = {}
total = 0

print("------ MENU ------")
for item, price in menu.items():
    print(item, ":", price)

while True:
    item_name = input("Enter item name (or 'done' to finish): ")

    if item_name.lower() == "done":
        break

    if item_name in menu:
        qty = int(input("Enter quantity: "))
        order[item_name] = qty
    else:
        print("Item not available!")

print("\n------ BILL ------")
for item, qty in order.items():
    price = menu[item] * qty
    total += price
    print(item, "x", qty, "=", price)

print("Total Bill =", total)
```

```python
# Restaurant Bill using Class

class Restaurant:

    def __init__(self):
        self.menu = {"Tea":20,"coffee":40,"Cold Coffee": 80,"Pizza": 200,"Burger": 120,"Pasta": 150}
        self.order = {}

    def show_menu(self):
        print("------ MENU ------")
        for item, price in self.menu.items():
            print(item, ":", price)

    def take_order(self):
        while True:
            item = input("Enter item name (or 'done'): ")

            if item.lower() == "done":
                break

            if item in self.menu:
                qty = int(input("Enter quantity: "))
                self.order[item] = qty
            else:
                print("Item not available!")

    def generate_bill(self):
        total = 0
        print("\n------ BILL ------")

        for item, qty in self.order.items():
            price = self.menu[item] * qty
            total += price
            print(item, "x", qty, "=", price)

        print("Total Bill =", total)

r = Restaurant()
r.show_menu()
r.take_order()
r.generate_bill()
```

```python
import sqlite3

try:
    # Connect to database
    conn = sqlite3.connect("restaurant.db")
    cursor = conn.cursor()

    # Create table
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS menu(
        item TEXT,
        price INTEGER
    )
    """)

    # Insert data
    cursor.execute("DELETE FROM menu")
    cursor.execute("INSERT INTO menu VALUES('Pizza',200)")
    cursor.execute("INSERT INTO menu VALUES('Burger',120)")
    cursor.execute("INSERT INTO menu VALUES('Pasta',150)")
    cursor.execute("INSERT INTO menu VALUES('Coffee',80)")

    conn.commit()

except Exception as e:
    print("Database Error:", e)

else:
    print("Database connected successfully!")

    order = {}
    total = 0

    cursor.execute("SELECT * FROM menu")
    items = cursor.fetchall()

    print("------ MENU ------")
    for item in items:
        print(item[0], ":", item[1])

    while True:
        item_name = input("Enter item name (or 'done'): ")

        if item_name.lower() == "done":
            break

        cursor.execute("SELECT price FROM menu WHERE item=?", (item_name,))
        result = cursor.fetchone()

        if result:
            qty = int(input("Enter quantity: "))
            order[item_name] = (result[0], qty)
        else:
            print("Item not available!")

    print("\n------ BILL ------")
    for item, data in order.items():
        price = data[0] * data[1]
        total += price
        print(item, "x", data[1], "=", price)

    print("Total Bill =", total)

finally:
    conn.close()
```

```python
# College Admission using Dictionary

students = {}

while True:
    name = input("Enter Student Name (or 'done' to stop): ")

    if name.lower() == "done":
        break

    try:
        age = int(input("Enter Age: "))
        marks = float(input("Enter 12th Marks (%): "))

        print("Available Streams: Science, Commerce, Arts")
        stream = input("Enter Stream: ")

        if marks >= 50:
            status = "Admitted"
            print("Admission Successful!\n")
        else:
            status = "Rejected"
            print("Admission Rejected (Minimum 50% required)\n")

        students[name] = {
            "Age": age,
            "Marks": marks,
            "Stream": stream,
            "Status": status
        }

    except:
        print("Invalid Input! Try again.\n")

print("\n----- Admission List -----")
for name, details in students.items():
    print(name, "->", details)
```

```python
import sqlite3

conn = sqlite3.connect("college.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS students(
    student_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    course TEXT,
    marks REAL
)
""")
conn.commit()


# ---------------- CLASS ----------------
class CollegeAdmission:

    # Dictionary of Courses with Minimum Marks
    courses = {
        "BCA": 50,
        "BBA": 55,
        "BSc": 60,
        "BA": 45
    }

    def __init__(self, name, course, marks):
        self.name = name
        self.course = course
        self.marks = marks

    def check_eligibility(self):
        try:
            if self.course not in self.courses:
                raise ValueError("Course not available")

            if self.marks < 0 or self.marks > 100:
                raise ValueError("Invalid marks")

            if self.marks >= self.courses[self.course]:
                return True
            else:
                return False

        except ValueError as e:
            print("Error:", e)
            return False

    def save_student(self):
        cursor.execute("INSERT INTO students(name,course,marks) VALUES (?,?,?)",
                       (self.name, self.course, self.marks))
        conn.commit()
        print("Student record saved in database!")


try:
    name = input("Enter student name: ")
    course = input("Enter course (BCA/BBA/BSc/BA): ")
    marks = float(input("Enter percentage marks: "))

    student = CollegeAdmission(name, course, marks)

    if student.check_eligibility():
        print("Admission Approved ")
        student.save_student()
    else:
        print("Admission Rejected ")

except Exception as e:
    print("Unexpected Error:", e)

finally:
    conn.close()
```

```python
# College Admission using Class

class College:

    def __init__(self):
        self.students = {}

    def take_admission(self):

        while True:
            name = input("Enter Student Name (or 'done' to stop): ")

            if name.lower() == "done":
                break

            try:
                age = int(input("Enter Age: "))
                marks = float(input("Enter 12th Marks (%): "))

                print("Available Streams: Science, Commerce, Arts")
                stream = input("Enter Stream: ")

                if marks >= 50:
                    status = "Admitted"
                    print("Admission Successful!\n")
                else:
                    status = "Rejected"
                    print("Admission Rejected (Minimum 50% required)\n")

                self.students[name] = {
                    "Age": age,
                    "Marks": marks,
                    "Stream": stream,
                    "Status": status
                }

            except:
                print("Invalid Input! Try again.\n")

    def show_students(self):
        print("\n----- Admission List -----")
        for name, details in self.students.items():
            print(name, "->", details)


# Creating object
c = College()
c.take_admission()
c.show_students()
```

```python
# Hotel bill printing

class Hotel:

    def __init__(self):
        self.total_amount = 0
        self.orders = []

    def take_order(self):
        count = int(input("Enter number of items: "))

        for i in range(1, count + 1):

            item_name = input("Enter item name: ")
            quantity = int(input("Enter quantity: "))
            price_per_item = int(input("Enter price: "))

            amount = quantity * price_per_item
            self.total_amount += amount

            self.orders.append((i, item_name, quantity, amount))

    def deliver_order(self):
        print("Order is ready!")

    def print_bill(self):

        print("=" * 40)
        print("{:^40}".format("Hotel Bill"))
        print("=" * 40)

        print("{:<3} {:<15} {:<5} {:<7}".format("No", "Item", "Qty", "Amount"))
        print("=" * 40)

        for record in self.orders:
            print("{:<3} {:<15} {:<5} {:<7}".format(record[0], record[1], record[2], record[3]))

        print("=" * 40)
        print("{:<25} {}".format("Total Amount", self.total_amount))
        print("=" * 40)


h = Hotel()

h.take_order()
h.deliver_order()
h.print_bill()
```

```python
class CollegeFee:

    def __init__(self):
        self.total = 0
        self.core_fee = 200000

    def calculate_fee(self):

        subject = input("Enter subject (HR/Finance/Marketing/DS): ")
        analytics = input("Do you want Analytics? (Y/N): ").upper()
        hostel = input("Hostel required? (Y/N): ").upper()
        food_months = int(input("How many months food needed?: "))
        transport = input("Transportation (semester/annual): ").lower()

        # Core fee
        self.total += self.core_fee

        # Analytics fee (10% extra)
        if analytics == "Y":
            self.total += self.core_fee * 0.10

        # Hostel fee
        if hostel == "Y":
            self.total += 200000

        # Food charges
        self.total += food_months * 2000

        # Transportation charges
        if transport == "semester":
            self.total += 13000
        elif transport == "annual":
            self.total += 13000 * 2

    def display_total(self):
        print("Total Annual Cost =", self.total)


# Create object
fee = CollegeFee()
fee.calculate_fee()
fee.display_total()
```
