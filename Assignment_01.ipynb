LIST COMPREHENSIONS

```python
# Q1: Write a list comprehension to generate squares of numbers from 1 to 10.
squares = [x**2 for x in range(1, 11)]
print(squares)
```

```python
# Q2: Create a list of even numbers between 1 and 50 using list comprehension.
even_numbers = [x for x in range(1, 51) if x % 2 == 0]
print(even_numbers)
```

```python
# Q3: Convert all strings in a list to uppercase using list comprehension.
words = ["apple", "banana", "cherry"]
uppercase_words = [word.upper() for word in words]
print(uppercase_words)
```

```python
# Q4: Given a list of integers, create a new list that contains only the positive numbers.
numbers = [-5, 10, -3, 8, 0, -2, 7]
positive_numbers = [num for num in numbers if num > 0]
print(positive_numbers)
```

```python
# Q5: Create a list of tuples (num, num^2) for numbers 1 to 5.
tuples_list = [(num, num**2) for num in range(1, 6)]
print(tuples_list)
```

```python
# Q6: Extract all vowels from a given string using list comprehension
text = "List Comprehension"
vowels = [char for char in text if char.lower() in "aeiou"]
print(vowels)
```

```python
# Q7: Flatten a 2D list using list comprehension.
matrix = [[1, 2], [3, 4], [5, 6]]
flattened = [num for row in matrix for num in row]
print(flattened)
```

```python
# Q8: Replace negative numbers with 0
numbers = [-5, 4, -2, 9, -1]
updated_list = [0 if num < 0 else num for num in numbers]
print(updated_list)
```

```python
# Q9: Length of each word
words = ["statistics", "python", "data", "analysis"]
lengths = [len(word) for word in words]
print(lengths)
```

```python
# Q10: Filter words starting with A or a
words = ["Apple", "banana", "Avocado", "grape", "apricot"]
filtered_words = [word for word in words if not word.lower().startswith("a")]
print(filtered_words)
```

```python
# Q11: Generate even/odd list
numbers = [1, 2, 3, 4, 5]
result = ["even" if num % 2 == 0 else "odd" for num in numbers]
print(result)
```

```python
# Q12: Divisible by both 3 and 5
divisible = [x for x in range(1, 101) if x % 3 == 0 and x % 5 == 0]
print(divisible)
```

```python
# Q13: Multiplication table (1â€“5)
table = [[i * j for j in range(1, 6)] for i in range(1, 6)]
print(table)
```

```python
# Q14: Convert dictionary keys to list
my_dict = {"name": "Dhanshree", "age": 22, "course": "MSc Statistics"}
keys_list = [key for key in my_dict]
print(keys_list)
```

```python
# Q15: Extract numeric digits
text = "Data123Science456"
digits = [char for char in text if char.isdigit()]
print(digits)
```

```python
# Q16: Remove spaces from string
text = "List Comprehension in Python"
no_spaces = [char for char in text if char != " "]
print("".join(no_spaces))
```

```python
# Q17: Characters appearing more than once
text = "programming"
duplicates = [char for char in set(text) if text.count(char) > 1]
print(duplicates)
```

```python
# Q18: Extract all words from list of sentences
sentences = ["Python is powerful", "Statistics is interesting"]
words = [word for sentence in sentences for word in sentence.split()]
print(words)
```

```python
# Q19: Unique elements from list
numbers = [1, 2, 2, 3, 4, 4, 5]
unique = [num for i, num in enumerate(numbers) if num not in numbers[:i]]
print(unique)
```

```python
# Q20: Cartesian product
A = [1, 2, 3]
B = ['a', 'b']
pairs = [(x, y) for x in A for y in B]
print(pairs)
```

LAMBDA FUNCTIONS

```python
from functools import reduce
import math
```

```python
# Q1: Add two numbers
add = lambda a, b: a + b
print(add(5, 3))
```

```python
# Q2: Check even
is_even = lambda x: x % 2 == 0
print(is_even(10))
```

```python
# Q3: Last character
last_char = lambda s: s[-1]
print(last_char("Python"))
```

```python
# Q4: Square using map
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, numbers))
print(squares)
```

```python
# Q5: Filter odd numbers
numbers = [1, 2, 3, 4, 5]
odd_numbers = list(filter(lambda x: x % 2 != 0, numbers))
print(odd_numbers)
```

```python
# Q6: Sort tuples by second value
data = [(1, 3), (2, 1), (4, 2)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

```python
# Q7: Check palindrome
is_palindrome = lambda s: s == s[::-1]
print(is_palindrome("madam"))
```

```python
# Q8: Maximum of three
max_three = lambda a, b, c: max(a, b, c)
print(max_three(10, 25, 15))
```

```python
# Q9: Reverse string
reverse = lambda s: s[::-1]
print(reverse("Dhanshree"))
```

```python
# Q10:
str_numbers = ["10", "20", "30"]
int_numbers = list(map(lambda x: int(x), str_numbers))
print(int_numbers)
```

```python
# Q11: Remove empty strings from list
strings = ["Python", "", "Data", "", "Science"]
non_empty = list(filter(lambda x: x != "", strings))
print(non_empty)
```

```python
from functools import reduce
# Q12: Factorial using lambda + reduce
number = 5
factorial = reduce(lambda x, y: x * y, range(1, number + 1))
print(factorial)
```

```python
# Q13: Larger of two numbers
larger = lambda a, b: a if a > b else b
print(larger(15, 20))
```

```python
# Q14: Check if number divisible by 5
divisible_by_5 = lambda x: x % 5 == 0
print(divisible_by_5(25))
```

```python
# Q15: Add 10 to each element
numbers = [5, 10, 15]
updated = list(map(lambda x: x + 10, numbers))
print(updated)
```

```python
# Q16: Sort list of dictionaries by age
people = [
    {"name": "Amit", "age": 25},
    {"name": "Riya", "age": 20},
    {"name": "Karan", "age": 30}
]

sorted_people = sorted(people, key=lambda x: x["age"])
print(sorted_people)
```

```python
# Q17: Check if character is vowel
is_vowel = lambda ch: ch.lower() in "aeiou"
print(is_vowel("A"))
```

```python
# Q18: Extract words with length > 5
words = ["Python", "Data", "Science", "AI", "MachineLearning"]
long_words = list(filter(lambda x: len(x) > 5, words))
print(long_words)
```

```python
import math
# Q19: Area of circle
area_circle = lambda r: math.pi * r**2
print(area_circle(7))
```

```python
# Q20: Remove duplicates using filter + set
numbers = [1, 2, 2, 3, 4, 4, 5]
unique = list(filter(lambda x: numbers.count(x) == 1, numbers))
print(unique)
```

```python
# Q21: Product of all numbers
numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

```python
# Q22: Absolute value
absolute = lambda x: x if x >= 0 else -x
print(absolute(-15))
```

```python
# Q23: Sort strings by length
words = ["Statistics", "AI", "Python", "Data"]
sorted_words = sorted(words, key=lambda x: len(x))
print(sorted_words)
```

```python
# Q24: Extract only uppercase characters
text = "PyTHon IS FuN"
uppercase_chars = list(filter(lambda x: x.isupper(), text))
print(uppercase_chars)
```

```python
# Q25: Square if even, cube if odd
check = lambda x: x**2 if x % 2 == 0 else x**3
print(check(4))
print(check(3))
```

```python
# Q26: Convert Celsius to Fahrenheit
celsius = [0, 20, 30, 40]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
```

```python
# Q27: Check anagrams
anagram = lambda s1, s2: sorted(s1.lower()) == sorted(s2.lower())
print(anagram("listen", "silent"))
```

```python
# Q28: Extract numeric values
mixed = [10, "Python", 25.5, "Data", 7]
numeric_values = list(filter(lambda x: isinstance(x, (int, float)), mixed))
print(numeric_values)
```

```python
# Q29: Check if any element is negative
numbers = [5, 10, -3, 8]
has_negative = any(map(lambda x: x < 0, numbers))
print(has_negative)
```

```python
# Q30: Generate multiplier function
multiply_by_n = lambda n: lambda x: x * n

double = multiply_by_n(2)
print(double(10))
```
