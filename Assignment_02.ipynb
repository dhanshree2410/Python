```python
# 15.
# Create FD interest bucket.
# Ask start and end date.
# Calculate number of days and assign interest rate.

from datetime import datetime

start_date = input("Enter start date (dd-mm-yyyy): ")
end_date = input("Enter end date (dd-mm-yyyy): ")

d1 = datetime.strptime(start_date, "%d-%m-%Y")
d2 = datetime.strptime(end_date, "%d-%m-%Y")

days = (d2 - d1).days

print("Total Days =", days)

if 7 <= days <= 45:
    rate = 5.75
elif 46 <= days <= 179:
    rate = 6.25
elif 180 <= days <= 210:
    rate = 6.35
elif 211 <= days < 365:
    rate = 6.40
elif 365 <= days < 730:
    rate = 7.00
elif 730 <= days < 1095:
    rate = 6.75
elif 1095 <= days < 1825:
    rate = 6.70
elif days >= 1825:
    rate = 6.60
else:
    rate = 0

print("Applicable Interest Rate =", rate, "%")
```

```python
# Q36
# List2 = [2,70,'work', para, 2.5, [1,2,3], (1,2), {1,2}, {1:'a', 2:'b'}, 3,10,302.5]
# Multiply each and every element from list2 by 2 and store answer in list3

para = "hello"

List2 = [2,70,'work', para, 2.5, [1,2,3], (1,2), {1,2}, {1:'a', 2:'b'}, 3,10,302.5]

list3 = []

for item in List2:
    if isinstance(item, (int, float, str, list, tuple)):
        list3.append(item * 2)
    elif isinstance(item, set):
        list3.append(item.union(item))  # set multiply concept
    elif isinstance(item, dict):
        new_dict = {}
        for k, v in item.items():
            new_dict[k] = v
            new_dict[k*2] = v
        list3.append(new_dict)

print("New list =", list3)
```

```python
# Q44.
# Create table cust_info with columns:
# sr_no, name, DOB, mobile
# Validate data and insert into table.

import sqlite3

# Connect database
conn = sqlite3.connect("customer.db")
cur = conn.cursor()

# Create table
cur.execute("""
CREATE TABLE IF NOT EXISTS cust_info(
    sr_no INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    DOB TEXT,
    mobile TEXT
)
""")

# Take input
name = input("Enter Name: ")
dob = input("Enter DOB (dd-mm-yyyy): ")
mobile = input("Enter Mobile Number: ")

# Validation
if name.isalpha() and mobile.isdigit() and len(mobile) == 10:
    cur.execute("INSERT INTO cust_info (name, DOB, mobile) VALUES (?, ?, ?)",
                (name, dob, mobile))
    conn.commit()
    print("Record Inserted Successfully ")
else:
    print("Validation Failed ")

conn.close()
```

```python
# Q56.
# Build function to print bill.

def print_bill(items):
    total = 0
    print("----- BILL -----")
    for item, price in items.items():
        print(item, ":", price)
        total += price
    print("----------------")
    print("Total Amount =", total)

cart = {
    "Pen": 20,
    "Book": 100,
    "Pencil": 10
}

print_bill(cart)
```
